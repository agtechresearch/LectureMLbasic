##  1. 파이썬 구조

▶ **표현식 - 문장 - 프로그램** <br>
> **표현식 (Expression)** : 파이썬에서 어떠한 값을 만들어내는 간단한 코드 <br>
>
> **문장 (Statement)** : 표현식이 하나이상 모인 경우, 그 자체로 어떠한 값을 만들어낼 수
> 없으면 문장이 아님   
>
> **프로그램 (Program)** : 문장이 모여서 형성된 것
>
> image.png



▶ **변수명명 규칙과 예약어**
> * **변수 (variable)** : 어떠한 값을 저장하는 메모리 공간(그릇)
>
>   * 변수 이름 = 값  
>   * 대입연산자(=)를 이용하여 변수에 값 할당  
>
> * **변수명 규칙**
>> 문자, 숫자, _(언더바)포함 가능, 하지만 **숫자로 시작하는것은 안된다.**
>>
>> 공백이나 다른 특수문자는 사용할 수 없다. ex) _python^^  
>>
>> 대 소문자 구분 (myVar와 MyVar는 다른 변수) <br>
>>
>>
>> **예약어**는 변수명으로 사용하면 안된다. <br>
>> * **파이썬의 예약어 (reserved words)** <br>
>> True, False, None, and, or, not, break, continue, return, if, else, elif, for, while, except, finally, global, import, try 등 <br>
>>
>> 대표적인 영문 명명방식(Naming convention)
>> * **Camel case** <br>
>> 단어사이에 공백없이, 낙타의 등모양처럼 뒤에 **연결되는 단어의 첫 글자를 대문자** 표기 <br>
>> 예시 : myVar, outputList, myBioStat
>>
>> * **Snake case** <br>
>> 단어사이에 **공백을 언더바(_)로 표기하여 연결**하는 방식 <br>
>> 예시 : my_var, output_list, my_bio_stat
>

▶ 문자열 포맷

> 다양한 문자열 출력을 위해서 사용  
>
> 문장 속에 변수를 포함시키는 여러가지 방법들

 **% 기호 사용하기**

 * "문자열" % 값

 * 문자열 부분에 어느 값이 들어가는지 지정(표시)

 > %s - 문자열  
 > %d - 정수  
 > %f - 실수

 ```
 print("안녕하세요 %s" % ("여러분"))
 ```



**{  }+format 사용하기**  

* 문장 내 괄호를 넣고, 괄호 속에 출력하려는 값 넣어주기  

* 순서대로 중괄호에 들어감

    ```
print("오늘의 음료는 {},{} 입니다".format(beverage1, beverage2))
```

**f-string 사용하기**  

* 문자열 앞에 f를 붙이고 문장 내 중괄호에 출력하려는 값 적기  
```
print(f"오늘의 음료는 {beverage1},{beverage2} 입니다")
```


````python
beverage1 = '아메리카노'
beverage2 = '딸기라떼'

print("{} 나왔습니다. 가격은 {}원 입니다.".format(beverage1,4500))
````
