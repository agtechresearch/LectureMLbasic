▶ 인덱싱(Indexing)과 슬라이싱(Slicing)

**인덱싱(Indexing)**

> 문자열, 리스트 등에 부여된 번호를 지정하여 가리킴
>
> **객체[index]**
>
> 인덱스는 0에서부터 시작한다.
>
> 인덱스의 끝에서부터 접근하려면 -1부터 시작한다.


**슬라이싱(Slicing)**

> 문자열이나 리스트의 일부분을 다른 문자열 또는 리스트로 수정하려면 슬라이스 방식을 사용하여 대입
>
> **[ start : stop : step ]** (step은 생략가능)
>
> 인덱스 값을 공백으로 두면 첫 항목부터, 혹은 끝 항목 까지의 의미를 가짐
>
>> [ 1 : 10 ] => 1번 항목부터 9번항목까지
>>
>> [ : 10] => 0번 항목부터 9번까지
>>
>> [ 1 : ] => 1번 항목부터 마지막 항목까지

```python
# 인덱싱
coffee = ['에스프레소','아메리카노','카페라떼','카푸치노']

# 에스프레소 출력하기
print(coffee[0])

# 카페라떼 출력하기
print(coffee[2])

# 카푸치노 출력하기
print(coffee[-1])
```

```python
# 인덱싱으로 값 바꾸기 - coffee 항목 중 원하는 커피를 콜드브루로 바꾸기
coffee[2] = '콜드브루'
print(coffee)
```

```python
# 슬라이싱
sports = ['풋살', '족구', '야구', '농구', '배구']

# 풋살, 족구, 야구 자리에 축구 대입하기
sports[0:3] = ['축구'] # 풋살(0)과 족구(1), 야구(2)에 축구가 대입됨
print(sports)
```

▶ 리스트 항목 추가 (insert(첨자, 항목), append(항목))

```python
coffee = ['에스프레소','아메리카노','카페라떼','카푸치노']

# 에스프레소 자리에 아이스아메리카노 추가하기
coffee.insert(0, '아이스아메리카노')
print(coffee)

# 리스트 coffee에 신메뉴 추가하기
coffee.append('말차라떼') # 리스트 끝 부분에 말차라떼 추가
print(coffee)
```

▶ 리스트 항목 삭제 (remove(항목), pop(첨자))
> **remove(항목)**
>
> 특정 **항목을 리스트에서 제거**
>
> **pop(첨자)**
>
> **첨자위치에 해당하는 항목이 제거**되며 제거된 값을 리턴한다
>
> 첨자를 명시하지 않으면 마지막 항목이 선택되어 제거됨

```python
coffee = ['에스프레소', '아이스아메리카노', '아메리카노', '카페라떼', '카푸치노', '말차라떼']

# 에스프레소를 제거하기 - remove 사용
coffee.remove('에스프레소')
print(coffee)

# 카푸치노를 제거하기 - pop 사용
coffee.pop(-2)
print(coffee)
```

## 2. 조건문( if )
▶ **주어진 조건(if)에 대한 결과(True, False)에 따라 다른 코드 실행**

▶ **if** - 만약 ~ 라면/**else** - 그렇지 않다면~/**elif** - 이것은 어떤가?

▶ if - else, if - elif - else 등 연결로 중첩가능

▶ 조건 결과에 따라 두 문장 이상 실행시 **들여쓰기 오류 주의**

image.png

```python
# 조건문으로 숫자 판별하기
a = 200

if a < 100:
  print("a 는 100보다 작다.")
  print("거짓이므로 이 문장은 출력되지 않음")
else:
  print("a 는 100보다 크다.")
  print("프로그램 끝")
```

▶ **if문의 응용**
> if문 조건에 리스트 및 in을 함께 사용하여, 리스트안에 특정요소가 있는지 True, False 판별 가능

```python
# 내 딸기가 잘 있을까
myList = ['딸기','수박','사과']

if '딸기' in myList:
  print("딸기가 myList안에 있습니다.")
```

```python
# 수박 선별기
'''
* 초고당도 수박은 12브릭스 이상
* 당도선별 수박은 12브릭스 미만
* 마트에 유통할 수 없는 수박은 9브릭스 미만
'''
watermelon = int(input("내 수박의 당도는:"))

if watermelon >= 12:
    print("초고당도 수박이다!")
elif 9 <= watermelon < 12:
    print("추가 당도 선별에 들어가야 합니다")
else:
    print("안돼! 팔 수 없는 수박")
```

## 3. 반복문 (for, while)
▶ **특정 조건에 따라 코드를 반복하여 실행**

▶ for문
> **정해져있는 시퀀스(반복 범위나 대상)의 항목 값으로 반복**을 실행
>
> 여러개의 값을 갖는 시퀀스에서 변수에 하나의 값을 순서대로 할당
>
> 할당된 변숫값을 갖고 블록의 문장들을 순차적으로 실행
>
> 반복과정을 **시퀀스의 마지막 항목까지 수행**
>
> image.png

```python
sum = 0

# 내장함수 range를 사용하여 0부터 9까지 i에 순차적으로 대입
for i in range(10):
    # 변수 sum에 i 더하기
    sum += i
    # 한 사이클을 돌 때마다의 i의 값과 sum의 값을 출력하기
    print("i :",i)
    print("sum :",sum)
else:
    # 0부터 9까지 다 돌고 나서의 결과 출력하기
    print(f"합:{sum}, 평균:{sum/10}")
```

```python
# for 문 안의 또 for문(중첩for문)

# 0에서 2까지 1씩 증가한 값을 i에 할당
for i in range(0, 3, 1):
    # 0에서 1까지 1씩 증가한 값을 k에 할당
    for k in range(0,2,1):
        print(f'반복문 실행 -> i:{i}, k:{k}')
```

▶ While 반복문
>  for와 유사한 반복을 수행
>
>  **조건식이 참이면 계속 반복**
>
> **일정 횟수이상 반복 후에도 종료조건을 만족시킬 수 없는 경우 무한루프에 빠지므로, 종료조건을 제대로 지정한 후 실행할 것**

image.png

```python
# 1에서 10까지 합계 구하기(while)
sum = 0
i = 1

# i가 11보다 작으면(미만) 반복하라
while i <11:
    # sum에 i 값을 더한다
    sum += i
    # 1씩 증가하는 숫자
    i += 1
    print(sum)
print("1에서 10까지의 합계 : %d"% sum)
```

▶ break과 continue문
> **break** <br>
> 반복문 실행 도중 break문을 만날경우, **나머지 반복문을 실행하지 않고 반복문 중지**
>
> **continue** <br>
> continue문을 만날 경우, **continue 이후의 반복 몸체를 실행하지 않고 다음 반복을 실행**
>
> 루프를 빠져나오지 않고 continue 아래의 문장만을 건너뛰는 역할
> 조건식이 거짓일때만 반복문 종료
>
> image.png

```python
# 1~100에서 3의배수만 제외하고 더하기
sum = 0

for i in range(1, 101):
    # i가 3의 배수일 때
    if i % 3 == 0:
        # 아래의 코드를 무시하고 다음 반복으로 넘어가라
        continue
    # 3의 배수만 제외하고 합하기
    sum += i

print("1~100에서 3의배수 제외한 합계: ", sum)
```

```python
# 파이썬으로 그림그리기 - * 을 이용해서 다음과 같이 출력하기(while 사용!)

"""
*
**
***
****
*****
"""

i = 1
while i <= 5:
    print("*" * i)
    i += 1
```
