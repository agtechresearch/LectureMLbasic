> Í∏∞Î≥∏Í∞í: `axis=0` (Ìñâ Î∞©Ìñ•, ÏúÑ/ÏïÑÎûò Î∞©Ìñ•ÏúºÎ°ú Ïó∞Í≤∞) <br>
> `axis=1` (Ïó¥ Î∞©Ìñ•, Ï¢å/Ïö∞ Î∞©Ìñ•ÏúºÎ°ú Ïó∞Í≤∞)
>
> * **.merge( Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ1, Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ2, on=Í≥µÌÜµÏª¨ÎüºÎ™Ö )** <br>
>  Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùÑ **Í≥µÌÜµÎêú Ìï≠Î™©(keyÍ∞í,Ïª¨Îüº)ÏùÑ Í∏∞Ï§ÄÏúºÎ°ú Ìï©Ïπ† Îïå** ÏÇ¨Ïö©(Î≥ëÌï©)

```python
# Îç∞Ïù¥ÌÑ∞ ÎßåÎì§Í∏∞
df1 = pd.DataFrame({
    'Ï±ÖÏù¥Î¶Ñ': ['Î™®Ïàú', 'Î™®ÎûòÏïåÎßå Ìïú ÏßÑÏã§Ïù¥ÎùºÎèÑ', 'ÏÇ¨ÌîºÏóîÏä§', 'ÎèàÏùò Ïã¨Î¶¨Ìïô', 'ÏõêÏîΩ'],
    'Ï∂úÌåêÏùº': ['2013-04-01', '2022-06-30', '2023-04-01', '2023-11-06', '2013-08-30'],
    'Î∂ÑÎ•ò': ['ÏÜåÏÑ§', 'ÏóêÏÑ∏Ïù¥', 'Ïù∏Î¨∏', 'Í≤ΩÏ†ú', 'ÏûêÍ∏∞Í≥ÑÎ∞ú']})
df1
```

```python
df2 = pd.DataFrame({
    'Ï±ÖÏù¥Î¶Ñ': ['ÏΩîÏä§Î™®Ïä§', 'Î™®ÎûòÏïåÎßå Ìïú ÏßÑÏã§Ïù¥ÎùºÎèÑ', 'Î™®Ïàú', 'Í≥µÏ†ïÌïòÎã§Îäî Ï∞©Í∞Å', 'ÏÇ¨ÌîºÏóîÏä§'],
    'Í∞ÄÍ≤©': ['19,900Ïõê', '17,000Ïõê', '13,000Ïõê', '18,000Ïõê', '26,800Ïõê']})
df2
```

* on = 'Í∏∞Ï§ÄÏó¥'

> Í≥µÌÜµÏù¥ ÎêòÎäî Í∏∞Ï§Ä Ïó¥Ïù¥ Ïó¨Îü¨Í∞úÏùº Îïå
>
> `pd.merge(df1, df2, on=['Ïó¥1','Ïó¥2']`
>
> Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏùò Í∏∞Ï§ÄÏù¥ ÎêòÎäî Ïó¥ Ïù¥Î¶ÑÏù¥ Îã§Î•º Îïå
>
> `pd.merge(df1, df2, left_on='ÏôºÏ™Ω Ïó¥', right_on='Ïò§Î•∏Ï™Ω Ïó¥')`

```python
# Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ìï©ÏπòÍ∏∞ - Ï±ÖÏù¥Î¶ÑÏùÑ Í∏∞Ï§ÄÏúºÎ°ú
pd.merge(df1,df2, on='Ï±ÖÏù¥Î¶Ñ')
```
